sweetviz is a great tool to quickly get a view of your dataset.  

```python
import pandas as pd
import numpy as np
import sweetviz as sv
```

```python
#read in data 
df = pd.read_csv("data/flight_and_dimension.csv", sep='\t',encoding='utf-8')
df.head()
```

```python
my_report = sv.analyze(df)
my_report.show_html()
```

```python
df.info()
```

## Fixing Type Column
There are some discs that have spaces after their disc type

```python
df['Type'].unique()
```

```python
#fixing the types with spaces at the end
df['Type'] = df['Type'].str.strip()
df['Type'].unique()
```

## Droping Unneeded Columns
drop In Prod., Date Approved, Certification Number

```python
df = df.drop(['In Prod.','Date Approved','Certification Number'],axis=1)
```

```python
df.head()
```

```python
my_report = sv.analyze(df)
my_report.show_html()
```
