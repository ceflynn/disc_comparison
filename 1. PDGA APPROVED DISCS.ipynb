## Remaining Mold Names

using the 'Original Mold', 'Brand', 'Date Approved' columns from the unmatched_df return all of the discs from the dimensions data that match the 'Brand' and 'Date Approved' columns then be able to select a matching disc if available.

```python
test_dict = {0:'A',1:'B'}
## adding the none of the above option to a dictionary
def none_of(dictionary):
    new_key = list(dictionary)[-1] + 1
    dictionary[new_key] = 'None Of The Above'
    return dictionary
none_of(test_dict)
```

```python
##function to return matching discs based on 'Brand', 'Date Approved' given a mold name
def find_molds(mold_name):
    #get the brand and the date
    disc_brand = unmatched_df.loc[unmatched_df['Original_Mold']==mold_name, 'Brand'].iat[0]
    disc_date = unmatched_df.loc[unmatched_df['Original_Mold']==mold_name, 'Date Approved'].array[0]
   
    #use brand and date to find all discs that match
    discs_df = dm_df[(dm_df['Brand']==disc_brand) & (dm_df['Date Approved']==disc_date)].reset_index(drop=True)
    disc_dict = dict(zip(discs_df.index, discs_df['Mold']))
    #add the "None of the Above" option and return
    choices =none_of(disc_dict)
    return choices
#test with a disc from unmatched_df above.
find_molds('Spring Ox')
```

## Function to Replace Mismatched Names

This function will go through each disc in the unmatched_df and provide options using the 'Brand' and 'Date Approved'and the functions we created above.  

```python
#updating the names in the dm_df with lambda function applied to our unmatched_df
def replace_mismatched(x):
    try:
        print(x)
        print("")
        dict2 = find_molds(x)
    except: 
        dict2 = {1:'Yes',2:'No'}
    response = dict_select(dict2)
    print("")
    return response

##apply to unmatched_df
unmatched_df['Mold_Change'] = unmatched_df['Original_Mold'].apply(lambda x: replace_mismatched(x))
```
