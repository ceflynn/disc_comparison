### Get the Table Headers

```python
table = souped.find('table',{"id":"table_1"})
header = table.findAll('th')
headers=[]
#don't include the options row
for h in header[:9]:
    headers.append(h.text)
headers
```

### Get the Data!
Now we need to start at spot 2 to get all of the disc data this will use td and create a list of lists for each 9 items.  There are 9 columns in the table.

```python
#list of all discs
disc_data=[]
#data with all the discs 
disc_text = table.findAll('td')
#empty temp list and counter starting at 0
counter = 0
temp_list=[]

#loop through each 9 cells and create a row for each disc.
for d in disc_text:
    d_text = d.text
    if counter <=8:
        temp_list.append(d_text)
        counter+=1
    else:
        counter=1
        disc_data.append(temp_list)
        temp_list=[]
        temp_list.append(d_text)

```

## Fixing Mold Names

1. Eliminate discs in flight_number_df that are not in our same_brands list

```python
#1
#only include the brands in our list.
fn_df = flight_number_df[flight_number_df['Brand'].isin(same_brands)].copy()
dm_df = dimensions_df.copy()
fn_df.shape
```

## difflib! 
difflib can be used to match items from one column with the best match from another column.  

```python
#2
#import difflib
import difflib
#add a duplicate column to keep track of original name
fn_df['Original_Mold'] = fn_df['Mold']
#convert the Brand in the fn_df dataset to closest match
fn_df['Mold'] = fn_df['Mold'].apply(lambda x: (difflib.get_close_matches(x, dm_df['Mold'])[:1] or [None])[0])
fn_df.head()
```

```python
#check the how many don't match with .shape
number = fn_df[fn_df['Mold'] != fn_df['Original_Mold']].shape[0]
print(f"There are {number} discs that do not match")
```

```python
#drop these discs for now.  
filtered_fn_df = fn_df[fn_df['Original_Mold'] == fn_df['Mold']]
new_number = filtered_fn_df.shape[0]
print(f"There are {new_number} discs that were matched correctly")
```
